
To learn more about the World Bank API, click [here](https://pypi.org/project/wbgapi/).

````python
import pandas as pd
import wbgapi as wb
````

Choose countries:

````python
wb.source.info()
````

````python
country_codes = ["GBR","FRA","HUN","PAK","SDN","SMR","THA","TSS","AGO","AFW"] # United Kingdom and France
````

Search for identifiers:

````python
wb.search("Climate change")
````

We are interested in:

- GDP (current US$)

- Population, total

````python
indicator_ids = ["SP.URB.TOTL.IN.ZS","NV.AGR.TOTL.ZS", "EG.ELC.ACCS.ZS","AG.LND.IRIG.AG.ZS","SP.URB.TOTL.IN.ZS","SP.URB.TOTL"]
````

Display identifiers info:

````python
wb.series.info(indicator_ids)
````

Get the data:

````python
# specicif time range
df = wb.data.DataFrame(indicator_ids, country_codes, time=range(2016, 2021))
df
````

````python
# most recent years
df =  wb.data.DataFrame(indicator_ids, country_codes, mrv=5)
df
````

````python
# Eliminating the NaN Values
df.dropna()
````

````python
# Statistical Description of the retrieved dataset
df.describe().T
````

````python
# Capturing the Time series dataset
import matplotlib.pyplot as plt
from matplotlib.pyplot import figure
figure(figsize = (10, 6))
wb.data.DataFrame('SP.URB.TOTL', economy=wb.region.members('SAS'), time=range (2016, 2020), numericTimeKeys=True, labels=True).set_index('Country').transpose().plot(title='Income growth in South Africa')
````

````python
wb.data.DataFrame('SP.URB.TOTL', ["GBR","FRA","HUN","PAK","SDN","SMR","THA","TSS","AGO","AFW"], time=range(2016,2020),numericTimeKeys=True).transpose().plot()
````

````python
# Creating the series dataframe
df_c = df.groupby('series')
df_c.head()
df_c.describe()
````
